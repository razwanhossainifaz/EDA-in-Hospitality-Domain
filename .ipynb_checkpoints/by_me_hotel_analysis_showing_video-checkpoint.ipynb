```python
df_agg_bookings["occ_pct"] = df_agg_bookings["successful_bookings"]/df_agg_bookings["capacity"]
df_agg_bookings.head()
```

```python
def actual_num(x):
    x = round((x * 100),2)  # aita ami nije think kore korsi
    return x

df_agg_bookings["occ_pct"] = df_agg_bookings["occ_pct"].apply(actual_num)
# df_agg_bookings["occ_pct"] = df_agg_bookings["occ_pct"].apply(lambda x : round(x*100, 2)) # ar aita sir er
df_agg_bookings.head()
```

**1. What is an average occupancy rate in each of the room categories?**

```python
df_agg_bookings.groupby("room_category")["occ_pct"].mean().round(2)
```

```python
df_rooms
```

```python
df = pd.merge(df_agg_bookings, df_rooms, left_on="room_category", right_on="room_id")
df.head()
```

```python
df.groupby(["room_category", "room_class"])["occ_pct"].mean().round(2)
```

```python
df.drop("room_id", axis=1, inplace=True)
df.head(4)
```

**2. Print Average occupancy rate per day**

```python
df_hotels
```

```python
df = pd.merge(df, df_hotels, on = "property_id")
df.head(4)
```

```python
df.groupby("city")["occ_pct"].mean().round(2).plot(kind = "bar")
```

**3. When was the occupancy better? Weekday or weekend?**

```python
df = pd.merge(df, df_date, left_on="check_in_date", right_on="date")
df.head(4)
```

```python
df.groupby("day_type")["occ_pct"].mean().round(2)
```
